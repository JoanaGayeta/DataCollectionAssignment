```python
import requests


URL="https://www.space.com/news/archive/2021/03"

page=requests.get(URL)
print(page.content)


from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

#Getting main contents
news_main = soup.find_all("ul", class_="day-list")
```

## March 11, 2021 News articles

```python
#Getting the contents of the news articles from March 11
news11 = news_main[7]
news11.contents
```

```python
#Getting the link of the articles
news11links = news11.find_all('a', href=True)

#Check if the links are in the list
news11links[0]
```

```python
#Open the link and get the ff: Title, Author, Date published, full Article

#9 articles
num_articles11= 9

print('\n+-------------------------------+')
print('|  March 11,2021 News Articles  |')
print('+-------------------------------+\n\n')

for n in range(0,num_articles11):
    #open link and get contents
    link1 = news11links[n].get('href')
    article1 = requests.get(link1)
    article_content1 = article1.content
    soup_article1 = BeautifulSoup(article_content1, 'html5lib')
    
    #get title
    title1 = soup_article1.find('h1').get_text()
    
    #get author
    author1 = soup_article1.find('span', style="white-space: nowrap").get_text()
    
    #date published
    date1 = soup_article1.find('time', itemprop="datePublished").get_text()
    
    #full article
    body1 = soup_article1.find_all(id="article-body")
    w = body1[0].find_all('p')
    
    # Unify all the paragraphs
    list_paragraphs1 = []
    for p in range(0, len(w)-1):
        paragraph1 = w[p].get_text()
        list_paragraphs1.append(paragraph1)
        fullarticle1 = "\n".join(list_paragraphs1)

    #print the articles
    print(f'\nArticle #', n+1)
    print('-------------')
    print(f'{title1}')
    print(f'  By:{author1}  | {date1}\n') 
    print(fullarticle1)
    
    
```

```python
import json

#json for march11 articles
march11_json =[]


for n in range(0,num_articles11):
    #open link and get contents
    link1 = news11links[n].get('href')
    article1 = requests.get(link1)
    article_content1 = article1.content
    soup_article1 = BeautifulSoup(article_content1, 'html5lib')
    
    #get title
    title1 = soup_article1.find('h1').get_text()
    
    #get author
    author1 = soup_article1.find('span', style="white-space: nowrap").get_text()
    
    #date published
    date1 = soup_article1.find('time', itemprop="datePublished").get_text()
    
    #full article
    body1 = soup_article1.find_all(id="article-body")
    w = body1[0].find_all('p')
    
    # Unify all the paragraphs
    list_paragraphs1 = []
    for p in range(0, len(w)-1):
        paragraph1 = w[p].get_text()
        list_paragraphs1.append(paragraph1)
        fullarticle1 = "\n".join(list_paragraphs1)
    
    #append the articles
    march11_json.append({
        "articlenum": n+1,
        "title" : title1,
        "author": author1,
        "date": date1,
        "fullarticle": fullarticle1
    })

#json for march 12 articles
march12_json=[]


for m in range(0,num_articles12):
    #open link and get contents
    link2 = news12links[m].get('href')
    article2 = requests.get(link2)
    article_content2 = article2.content
    soup_article2 = BeautifulSoup(article_content2, 'html5lib')
    
    #get title
    title2 = soup_article2.find('h1').get_text()
    
    #get author
    author2 = soup_article2.find('span', style="white-space: nowrap").get_text()
    
    #date published
    date2 = soup_article2.find('time', itemprop="datePublished").get_text()
    
    #full article
    body2 = soup_article2.find_all(id="article-body")
    z = body2[0].find_all('p')
    
    # Unify all the paragraphs
    list_paragraphs2 = []
    for k in range(0, len(z)-1):
        paragraph2 = z[k].get_text()
        list_paragraphs2.append(paragraph2)
        fullarticle2 = "\n".join(list_paragraphs2)
    
    #print the articles
    march12_json.append({
        "articlenum": m+1,
        "title" : title2,
        "author": author2,
        "date": date2,
        "fullarticle": fullarticle2
    })
    

articles_json=[]

articles_json.append(march11_json)
articles_json.append(march12_json)

with open('news_articles.json', 'w') as json_file:
    json.dump(articles_json, json_file)

articles_json
    
```
